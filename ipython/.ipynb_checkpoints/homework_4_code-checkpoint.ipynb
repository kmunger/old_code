<p>Load in both training and test sets. Note that each dataset is comman delimited, has a row header and the first field (eventid) should be used as an index.

Clean data:
<ul>
<li>Convert Label to numeric (choose the minority class to be = 1)</li>
    <ul>
    <li> Create a feature Y with numeric label</li>
    <li> Drop the feature "Label"</li>
    </ul>
<li>If a feature has missing values (-999):
    <ul>
    <li>Create a dummy var for the missing value</li>
    <ul><li>Call the variable "orig_var_name"+"mv" where "orig_var_name" is the name of the actual var with a missing value</li>
        <li>Give this new variable a 1 if the orig var is missing</li></ul>
    <li>Replace the missing value with the average of the feature (make sure to compute the mean where the value isn't missing.)</li>
    </ul>
</li>
</ul>

</p>

```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import course_utils as bd
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing 



from sklearn import linear_model
reload(bd)

f_train = '/Users/briand/Desktop/ds course/datasets/boson_training_cut_2000.csv'
f_test = '/Users/briand/Desktop/ds course/datasets/boson_testing_cut.csv'

def cleanBosonData(infile, scale):
    dat = pd.read_csv(infile, header = 0, sep = ',', index_col = 0)
    Y = (dat['Label'] == 's')*1
    dat = dat.drop('Label', 1)
    flds = dat.columns.values
    mv = -999.0
    #Loop through the fields and replace mv with avg and create dummy vars
    for f in flds:
        #Don't do this for Y or if no missing vals
        if (sum((dat[f]==mv))>0 and f !='Y'):
            mu_f = dat[f][(dat[f] != mv)].mean()
            dat[f+"mv"] = (dat[f] != mv)*1
            dat[f] = dat[f].replace(to_replace=mv, value=mu_f)
    if (scale==1):
        flds2 = dat.columns        
        scl = preprocessing.StandardScaler()
        sdat = pd.DataFrame(scl.fit_transform(dat), columns = flds2)
        sdat['Y'] = Y.values
        return sdat
    else:
        dat['Y'] = Y
        return dat


train = cleanBosonData(f_train, 1)
test = cleanBosonData(f_test, 1)
```

<p>Build out of the box models</p>

```python
lab = 'Y'
logreg = linear_model.LogisticRegression(C=1e30)
logreg.fit(train.drop(lab,1), train[lab])
p_lr = logreg.predict_proba(test.drop(lab,1))[:,1]

my_svm = svm.SVC(kernel='linear')
my_svm.fit(train.drop(lab,1), 2*train[lab]-1)
p_svm = my_svm.decision_function(test.drop(lab,1))
```

```python
def getAUC(truth, pred):
    fpr, tpr, thresholds = roc_curve(truth, pred)
    return auc(fpr, tpr)


def plotAUC(truth, pred, lab):
    fpr, tpr, thresholds = roc_curve(truth, pred)
    roc_auc = auc(fpr, tpr)
    c = (np.random.rand(), np.random.rand(), np.random.rand())
    plt.plot(fpr, tpr, color=c, label= lab+' (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC')
    plt.legend(loc="lower right")
    
```

<p>Now plot the AUCs</p>

```python
plotAUC(test['Y'], p_lr, 'LR')
plotAUC(test['Y'], p_svm, 'SVM')
plt.show()
```

```python
from sklearn.cross_validation import *

def xValSVM(tr, lab, k, cs):
    '''
    Perform k-fold cross validation on SVM across a range of C,
    returns mean and var of auc avg.
    '''
    n_samp = tr.shape[0]
    cv = KFold(n=n_samp, n_folds = k)
    aucs = {}

    for train_index, test_index in cv:
        tr_k = tr.iloc[train_index]
        va_k = tr.iloc[test_index]
    
        for c in cs:
            svm_k = svm.SVC(kernel='linear', C=c)
            svm_k.fit(tr_k.drop(lab,1), tr_k[lab])
            met = roc_auc_score(va_k[lab], svm_k.decision_function(va_k.drop(lab,1)))
            
            if (aucs.has_key(c)):
                aucs[c].append(met)
            else:
                aucs[c] = [met]
    
    return aucs
```

```python
xval_dict = {'e':[], 'mu':[], 'sig':[]}
k = 10
er = range(-9,1)
cs = [10**i for i in er ]
aucs_sv = xValSVM(train, 'Y', k, cs)

for i in er:
    xval_dict['e'].append(i)
    xval_dict['mu'].append(np.array(aucs_sv[10**i]).mean())
    xval_dict['sig'].append(np.sqrt(np.array(aucs_sv[10**i]).var()/10))
```

<p>Now retrain the SVM on the optimal C's</p>

```python
means = np.array(xval_dict['mu'])
cs = np.array(xval_dict['e'])
stderr = np.array(xval_dict['sig'])

low = (means-stderr).max()
best_c_ser = np.array(cs)[(means>low)].min()
best_c_max = np.array(cs)[(means==max(means))]

plt.plot(xval_dict['e'], means)
plt.plot(xval_dict['e'], means+stderr, 'k+-')
plt.plot(xval_dict['e'], means-stderr, 'k--')
plt.plot(xval_dict['e'], low*np.ones(len(means)), 'r')
```

```python
lab = 'Y'
svm_ser = svm.SVC(kernel='linear', C = 10**best_c_ser)
svm_ser.fit(train.drop(lab,1), 2*train[lab]-1)
p_svm_ser = svm_ser.decision_function(test.drop(lab,1))

svm_max = svm.SVC(kernel='linear', C = 10**best_c_max)
svm_max.fit(train.drop(lab,1), 2*train[lab]-1)
p_svm_max = svm_ser.decision_function(test.drop(lab,1))
```

```python
def modBootstrapper(train, test, nruns, sampsize, lr, c):
    '''
    Samples with replacement, runs multiple train/eval attempts
    returns mean and stdev of AUC
    '''
    lab = 'Y'
    auc_res = []
    for i in range(nruns):
        train_samp = train.iloc[np.random.randint(0, len(train), size=sampsize)]
        if (lr == 1):
            lr_i = linear_model.LogisticRegression(C=1e30)
            lr_i.fit(train_samp.drop(lab,1),train_samp[lab])
            p = lr_i.predict_proba(test.drop(lab,1))[:,1]
        else:
            svm_i = svm.SVC(kernel='linear', C=10**best_c_ser)
            svm_i.fit(train_samp.drop(lab,1),train_samp[lab])
            p = svm_i.decision_function(test.drop(lab,1))
        
        auc_res.append(roc_auc_score(test[lab], p))

    return [mean(auc_res),math.sqrt(var(auc_res)/nruns)]
```

```python
#Get LR Bootstrap

ns = [50, 100, 200, 500, 1000, 2000]

mu_lr = []; ser_lr = []
mu_mm = []; ser_mm = []
for n in ns:
    mu1, ser1 = modBootstrapper(train, test, 20, n, 1, 10**best_c_ser)
    mu_lr.append(mu1)
    ser_lr.append(ser1)
    mu2, ser2 = modBootstrapper(train, test, 20, n, 0, 10**best_c_ser)
    mu_mm.append(mu2)
    ser_mm.append(ser2)
    
    
```
